## Importación de librerías

```python
#Trataiento de datos.
import pandas as pd
pd.set_option('display.max_columns', None)
#--------------

#Exploración de archivos
import sys
sys.path.append('../')
#--------------

#Funciones personalizadas
from src import sp_union as su
#--------------
```

## Carga de archivos

```python
# Importación de archivo amazon_sales

amazon_sales = pd.read_excel('../data/amazon_sales.xlsx')

su.union_archivos(amazon_sales)
```

```python
# Importación de archivo amazon_customers

amazon_customers = pd.read_excel('../data/amazon_customers.xlsx')

su.union_archivos(amazon_customers)
```

## Unión de tablas

```python
# Unión de ambas tablas por las columnas comunes 'Customer_ID' y 'customer_id'
# Al integrar ambos conjuntos de datos mediante un 'inner join' no serán incluidos en el análisis los clientes que no hayan realizado pedidos.

data_raw = amazon_customers.merge(amazon_sales, left_on='customer_id', right_on='Customer_ID', how='inner')

su.union_archivos(data_raw)
```

```python
# Borrado de índice 'Row_ID' y de columna 'Customer_ID' sobrante.

data_raw.drop(columns=['Row_ID', 'Customer_ID'], inplace=True)

data_raw.sample(5)
```

```python
# Exportación de dataframe

data_raw.to_csv('../data/data_raw.csv', index=False)
```
