## Limpieza de datos

```python
# Conversión de nobmres de columnas a minúscula
df.columns = df.columns.str.lower()
```

```python
# Conversión de columnas order_date y signup_date a formato datetime
df['order_date'] = pd.to_datetime(df['order_date'], format = '%Y-%m-%d')
df['signup_date'] = pd.to_datetime(df['signup_date'], format = '%Y-%m-%d')
```

```python
# Creación de columnas de order_year y order_month para luego analizar los pedidos por año

df['order_year'] = df['order_date'].dt.year
df['order_month']  = df['order_date'].dt.month
```
