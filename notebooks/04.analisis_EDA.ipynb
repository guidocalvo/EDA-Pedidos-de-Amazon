```python
# Se comprueba cuales son las fechas más antiguas de los estados 'Preparing order' y 'Dispatched to center'.

fecha_mas_antigua_preparing = df.loc[df['state'] == 'Preparing order', 'order_date'].min()

fecha_mas_antigua_dispatched = df.loc[df['state'] == 'Dispatched to center', 'order_date'].min()

print(fecha_mas_antigua_preparing)
print(fecha_mas_antigua_dispatched)
```

```python
# Análisis de pedidos cancelados.
# Mapa de correlación de variables numéricas.

df_mapa = df.copy()

#Creamos una columna que se llame 'entregado' y tenga los valores 1 y 0. (Entregado = 1, Cancelado = 0)
df_mapa['entregado'] = df_mapa['state'].map({'Received by the customer': 1, 'Cancelled': 0,})

# Como los pedidos con estado 'Preparing order' y 'Dispatched to center' tienen una antiguedad máxima a tres meses se eliminan.
# Se analizan los pedidos entregados o cancelados de los últimos 4 años.
df_mapa = df_mapa.dropna(subset=['entregado'])


df_mapa['entregado'] = df_mapa['entregado'].astype('Int64')

cols_numericas = [col for col in df_mapa.select_dtypes(include=['float64','Int64']).columns]
block = df_mapa[cols_numericas]
mapa = block.corr()

mask = np.triu(np.ones_like(mapa, dtype=bool)) # Máscara para solo mostrar una mitad del gráfico.

plt.figure(figsize=(10,8))
sns.heatmap(
    mapa,
    mask = mask,
    annot=True,
    fmt='.2f',
    vmin=-1, vmax=1,
    cmap='coolwarm',
    square=True,
    linewidths=.5,
    annot_kws={'fontsize': 6},
    cbar_kws={'shrink': 0.5} 
)
plt.title('Correlation map')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de los pedidos completados por año.

fig, ax = plt.subplots(figsize=(12, 5))

# Se crea dataframe únicamente con los pedidos entregados.
df_completado = df_mapa[df_mapa['entregado'] == 1]

sns.countplot(
    data=df_completado,
    x='order_year',
    ax=ax
)

ax.set_title('Pedidos completados por año')
ax.set(xlabel=None, ylabel=None)

plt.tight_layout()

plt.show()
```

```python
# Gráfico de comparación entre pedidos con descuento y total de pedidos.

# Contar pedidos
total_pedidos = len(df)
pedidos_con_descuento = len(df[df['discount'] > 0])

# Crear DataFrame temporal para graficar
df_plot = pd.DataFrame({
    'Tipo Pedido': ['Con Descuento', 'Total'],
    'Cantidad': [pedidos_con_descuento, total_pedidos]
})

# Gráfico
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(data=df_plot, x='Tipo Pedido', y='Cantidad', ax=ax)

ax.set_title('Pedidos con descuento vs total de pedidos')
ax.set(xlabel=None, ylabel=None)

plt.tight_layout()
plt.show()
```

```python
# Gráfico de número de pedidos con un descuento mayor a 0.

# Filtrar descuentos mayores a 0
df_discount = df_completado[df_completado['discount'] > 0]

# Contar pedidos por nivel de descuento
discount_counts = (
    df_discount
    .groupby('discount')
    .size()
    .sort_index()
)

# Gráfico
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(
    discount_counts.index,
    discount_counts.values,
    marker='o',
    color='seagreen',
    label='Number of Orders'
)

ax.set_title('Number of Orders by Discount Level (>0)')
ax.set_xlabel('Discount')
ax.grid(True)
ax.legend()

plt.tight_layout()
plt.show()
```
